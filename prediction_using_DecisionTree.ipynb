# Prediction using Decision Tree 

```python
#creating a sample dataset 
#creating a dataset which includes the size and shade and according to it 
# predicting if it is an apple or a an orange 
#size is in cm 
#shade is between (1-10)
```

```python
x=[
    [7,2],
    [8,3],
    [9,8],
    [10,9]
]
y = [0,0,1,1] # 0 = apple , 1 = orange
```

```python
#loading the model 
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
#fitting data to the model
model.fit(x,y)
size = float(input("enter the size of fruit in cm:"))
shade = float(input("enter the shade of fruit between (1-10)"))
prediction =model.predict([[size,shade]])

if prediction==0:
    print("it is likely to be an apple")
else:
    print("it is likely to be an orange")
```
