## üî¢Converting Text Data to Numeric Values 


```python
cols =['Course_Category','Learning_Mode','Country','Completion_Status']
cols
```

```python
#using pandas get_dummies function which seperates the different row data into different column 
df_label = pd.get_dummies(df_label , columns = cols,dtype=int)
df_label
```

```python
df_label.drop(['Student_ID','Completion_Status_In Progress','Learning_Mode_Blended'] , axis=1,inplace = True)
```

## üìàStandardzing the data 
- Using the sklearn library to stardardize the data 


```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
cols1 = ['Video_Watch_Time_Min','Quiz_Score','Assignments_Completed','Login_Frequency','Time_Spent_Per_Session_Min']
```

```python
df_label[cols1] = scaler.fit_transform(df_label[cols1])
df_label
```

```python
#visualising data after standardising the data 
```

```python
sns.heatmap(df_label[selected_vars].corr(), annot=True)
plt.show()
```

# üß™Traning and Testing the Model 

```python
from sklearn.model_selection import train_test_split
x = df_label.drop(['Completion_Status_Completed'],axis=1)
y = df_label[['Completion_Status_Completed']]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, train_size = 0.7 , test_size=0.3 , random_state=100)
```

from sklearn.tree import DecisionTreeClassifier

## üßë‚ÄçüîßTraning the Model 
- Fitting the data to the model 

```python
model.fit(x_train , y_train)
```
